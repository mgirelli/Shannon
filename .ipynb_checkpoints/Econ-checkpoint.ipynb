```python
import pandas as pd
import zipfile
from pathlib import Path
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
from Shannon import entropic_bipartite
```

```python
zip_path = Path("BACI_HS92_V202601.zip")
out_dir  = Path("data/raw")

if not out_dir.exists() or not any(out_dir.glob("*.csv")):
    out_dir.mkdir(parents=True, exist_ok=True)
    with zipfile.ZipFile(zip_path, "r") as zf:
        zf.extractall(out_dir)
else:
    pass
```

```python
with zipfile.ZipFile("BACI_HS92_V202601.zip") as z:
    for name in z.namelist():
        print(name)
```

```python
with zipfile.ZipFile("BACI_HS92_V202601.zip") as z:
    # Anno campione
    with z.open("BACI_HS92_Y2015_V202601.csv") as f:
        df = pd.read_csv(f, dtype={'k':'str','i':'str' }) 
        print(df.head())
    
    # Codici paese
    with z.open("country_codes_V202601.csv") as f:
        countries = pd.read_csv(f, dtype='str')  
        print(countries.head())
    
    # Codici prodotto
    with z.open("product_codes_HS92_V202601.csv") as f:
        products = pd.read_csv(f, dtype= 'str')  
        print(products.head())
```

```python
df = df[["i","k","v"]]
countries = countries[["country_code","country_iso3"]]
df = df.merge(countries, left_on="i", right_on="country_code")
df = df.merge(products, left_on="k", right_on="code")
```

```python
#scelgo singolo prodotto
prodotto = "010111"
df_prod = df[df["k"] == prodotto ]
df_grouped = df_prod.groupby("country_iso3")["v"].sum().reset_index()
```

```python
url = "https://naturalearth.s3.amazonaws.com/110m_cultural/ne_110m_admin_0_countries.zip"

world = gpd.read_file(url)

gdf = world.merge(
    df_grouped,
    left_on="ADM0_A3",
    right_on="country_iso3",
    how="left"
)

gdf["log_v"] = np.log1p(gdf["v"])

gdf.plot(column="log_v",cmap="OrRd", legend=True, missing_kwds={"color": "lightgrey"})
plt.title(products.loc[products["code"] == prodotto, "description"].values[0])
plt.show()
```
