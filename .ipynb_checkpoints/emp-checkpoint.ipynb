```python
import biom
import pandas as pd
import numpy as np
import biom_to_csv as btc
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load dataframe
XSB_90bp    = btc.biom_to_csv('biom_data/emp_deblur_90bp.release1.biom')
#XSB_cr_13_8 = btc.biom_to_csv('biom_data/emp_cr_gg_13_8.release1.biom')
#XSB_silva   = btc.biom_to_csv('biom_data/emp_cr_silva_16S_123.release1.biom')
#XSB_100bp   = btc.biom_to_csv('biom_data/emp_deblur_100bp.release1.biom')
#XSB_150bp   = btc.biom_to_csv('biom_data/emp_deblur_150bp.release1.biom')
#XSB_gg_13_8 = btc.biom_to_csv('biom_data/emp_or_gg_13_8.no_sample_singletons.biom') #dataset molto grande
```

```python
Hp, Hs = btc.entropy(XSB_90bp)
print('Entropia phyla')
print(Hp)
print('Entropia sample')
print(Hs)
```

```python
meta = pd.read_csv('biom_data/emp_qiime_mapping_release1.tsv', sep='\t', index_col=0, low_memory=False)
meta_filtered = meta.loc[meta.index.isin(XSB_90bp.index)]
meta_aligned = meta_filtered.reindex(XSB_90bp.index)
```

```python
gruppi = {}
for gruppo in meta_aligned['empo_2'].unique():
    mask = meta_aligned['empo_2'] == gruppo
    gruppi[gruppo] = XSB_90bp[mask]
    print(f"{gruppo}: {gruppi[gruppo].shape}")
gruppi.pop("Negative")
gruppi.pop("Positive")
```

```python
fig, axes = plt.subplots(2, 3, figsize=(20, 12))
axes = axes.flatten()

for idx, (gruppo, df) in enumerate(gruppi.items()):
    # 1. Binarizza
    df_bin = (df > 0).astype(int)
    
    # 2. Calcola somme
    phyla_sum = df_bin.sum(axis=1)
    samples_sum = df_bin.sum(axis=0)
    
    # 3. Riordina
    df_sorted = df_bin.loc[
        phyla_sum.sort_values(ascending=False).index,
        samples_sum.sort_values(ascending=False).index
    ]
    
    # 4. Plot
    sns.heatmap(df_sorted,
                cmap='magma',
                cbar=False,
                xticklabels=False,
                yticklabels=False,
                ax=axes[idx])
    
    axes[idx].set_title(f'{gruppo} (n={df.shape[0]})', fontsize=13)
    axes[idx].set_xlabel('Campioni (ricchezza decrescente)')
    axes[idx].set_ylabel('Phyla (ubiquità decrescente)')

# Nascondi assi vuoti se i gruppi sono meno di 6
for idx in range(len(gruppi), len(axes)):
    axes[idx].set_visible(False)

plt.suptitle('Struttura Nested per gruppo ambientale (empo_2)', fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```
