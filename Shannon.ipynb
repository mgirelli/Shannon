```python
import pandas as pd
import zipfile
from pathlib import Path
```

```python
zip_path = Path("BACI_HS92_V202601.zip")
out_dir  = Path("data/raw")

if not out_dir.exists() or not any(out_dir.glob("*.csv")):
    out_dir.mkdir(parents=True, exist_ok=True)
    with zipfile.ZipFile(zip_path, "r") as zf:
        zf.extractall(out_dir)
else:
    pass
```

```python
with zipfile.ZipFile("BACI_HS92_V202601.zip") as z:
    for name in z.namelist():
        print(name)
```

```python
with zipfile.ZipFile("BACI_HS92_V202601.zip") as z:
    # Anno campione
    with z.open("BACI_HS92_Y2015_V202601.csv") as f:
        print(pd.read_csv(f, nrows=5).to_string())
    
    # Codici paese
    with z.open("country_codes_V202601.csv") as f:
        print(pd.read_csv(f, nrows=5).to_string())
    
    # Codici prodotto
    with z.open("product_codes_HS92_V202601.csv") as f:
        print(pd.read_csv(f, nrows=5).to_string())
```
